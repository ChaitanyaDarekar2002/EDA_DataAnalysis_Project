```python
data.drop_duplicates(inplace=True)
```

```python
data["Total_Expense"] = (
    data["Food_Expense_INR"] +
    data["Rent_Expense_INR"] +
    data["Transport_Expense_INR"] +
    data["Entertainment_Expense_INR"]
)
```

```python
print("Total_Expense")
```

```python
data["Savings_Amount"] = (
    data["Monthly_Income_INR"] * data["Savings_Rate_Pct"] / 100
)
```

```python
plt.figure(figsize=(8,5))
plt.hist(data["Age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
data["Gender"].value_counts().plot(kind="bar", figsize=(6,5))
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
data["City"].value_counts().plot(kind='line')
```

```python
city_income = data.groupby("City")["Monthly_Income_INR"].mean()
city_income.plot(kind="bar", figsize=(8,5),color='red')
plt.title("Average Income by City")
plt.ylabel("Income")
plt.show()
```

```python
data["BMI_Category"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(7,7)
)

plt.title("BMI Category Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    data["Monthly_Income_INR"],
    data["Savings_Rate_Pct"]
)

plt.title("Income vs Savings Rate")
plt.xlabel("Income")
plt.ylabel("Savings Rate")
plt.show()
```

```python
plt.bar(
    data["Marital_Status"],
    data["Age"]
)
plt.show()
```

```python
plt.pie(
    data["Age"]
)
plt.show()
```

```python
data["Smoking_Status"].value_counts().plot(
    kind="bar",
    figsize=(7,5)
)

plt.title("Smoking Status")
plt.ylabel("Count")
plt.show()
```

```python
data["Exercise_Frequency"].value_counts().plot(
    kind="bar",
    figsize=(7,5)
)

plt.title("Exercise Frequency")
plt.ylabel("Count")
plt.show()
```
